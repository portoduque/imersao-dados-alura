### Desafio 06: Renomear as colunas tirando o hífen


```python
old_names = dados.columns

new_names = [name.replace("-", "") for name in old_names]

dados.columns = new_names

dados

```

### Desafio 07: Deixar os gráficos bonitões. (matplotlib.pyplot)


```python
import matplotlib.pyplot as plt

tratamentos_data = dados["tratamento"].value_counts()
my_labels = ["com droga", "com controle"]
my_explode = [0, 0.3]
sizes = [50, 60]
my_colors = ["g", "hotpink"]

plt.title("Proporção das classes tratamento")
plt.pie(tratamentos_data, labels= my_labels, explode= my_explode, shadow= True, colors = my_colors, autopct='%1.1f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt

dados["tempo"].value_counts().plot.barh(title= "Tempos de avaliação dos efeitos", ylabel= "Frequencia", xlabel="Horas")

```
